# Lemmatization
* The lemma of 'was' is 'be' and the lemma of 'mice' is 'mouse'.
* The lemma of 'meeting' might be 'meet' or 'meeting' depending on its use in a sentence.

```python
# Perform standard imports:
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Function that return the Lemma table
def lemma_table(doc):
  t = PrettyTable(["Text", "Part of Speech", "Lemma Id", "Lemma"])

  for token in doc:
    t.add_row([token.text, token.pos_, token.lemma, token.lemma_])

  print(t)
```

```python
doc1 = nlp(u"I am a runner running in a race because I love to run since I ran today")

lemma_table(doc1)
```

<font color=lightgreen>In the above sentence, `running`, `run` and `ran` all point to the same lemma `run` (...11841) to avoid duplication.</font>

```python
doc2 = nlp(u"I saw eighteen mice today!")

lemma_table(doc2)
```

<font color=lightgreen>Notice that the lemma of `saw` is `see`, `mice` is the plural form of `mouse`, and yet `eighteen` is its own number, *not* an expanded form of `eight`.</font>

```python
doc3 = nlp(u"I am meeting him tomorrow at the meeting.")

lemma_table(doc3)
```

<font color=lightgreen>Here the lemma of `meeting` is determined by its Part of Speech tag.</font>

```python
doc4 = nlp(u"That's an enormous automobile")

lemma_table(doc4)
```

<font color=lightgreen>Note that lemmatization does *not* reduce words to their most basic synonym - that is, `enormous` `doesn't` become `big` and `automobile` `doesn't` become `car`.</font>
