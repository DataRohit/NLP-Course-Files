## Working with PDF Files
#### Q4: Use PyPDF2 to open the file `Business_Proposal.pdf`. Extract the text of page 2.

```python
!pip install PyPDF2
import PyPDF2
```

```python
business_proposal = open(base_path + "Business_Proposal.pdf", "rb")

pdf_reader = PyPDF2.PdfFileReader(business_proposal)

page_two_text = pdf_reader.getPage(1).extractText()

business_proposal.close()

print(page_two_text)
```

#### Q5: Open the file `contacts.txt` in append mode. Add the text of page 2 from above to `contacts.txt`.

#### CHALLENGE: See if you can remove the word "AUTHORS:"

```python
f = open(base_path + "contacts.txt", "a+")

f.write(page_two_text[8:])

f.seek(0)

file_text = f.read()

f.close()

print(file_text)
```

## Regular Expressions
#### Q6: Using the `page_two_text` variable created above, extract any email addresses that were contained in the file `Business_Proposal.pdf`.

```python
import re

pattern = r"[\w]+@ourcompany.com"

re.findall(pattern, page_two_text)
```
